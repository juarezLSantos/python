```python
lista = []
listaConcluida = []

def exibir_menu():
    #exibição do menu
    print("Menu:")
    print("1. Adicionar uma nova tarefa: ")
    print("2. mostrar tarefas existentes: ")
    print("3. Remover uma tarefa: ")
    print("4. Remover tarefa concluída: ")
    print("5. Marcar tarefa como Concluida: ")
    print("6. Sair")

def adicionar_tarefa():
    # adição de uma nova tarefa na lista
    lista.append(input("Adicione uma Tarefa"))
    print(lista)
    print("Nova tarefa adicionada!")

def remover_tarefas():
    # remoção das tarefas na lista

    try:
      lista.remove(input("Remova uma Tarefa"))
      print(lista)
    except TypeError:
      print("Escreveu Errado")

def remover_tarefas_concluidas():
    # remoção das tarefas concluidas na lista-concluida

    try:
      listaConcluida.remove(input("Remova uma Tarefa"))
      print(listaConcluida)
    except TypeError:
      print("Escreveu Errado")


def exibir_tarefas():
    # Exibição das listas
    print("Lista de tarefas:" )
    print(lista)
    print("Tarefas concluídas: ")
    print(listaConcluida)

def concluir_tarefa():
    concluido = input("Marcar como concluído")

    # remove tira da lista
    # append bota na lista_concluida
    lista.remove(concluido)
    listaConcluida.append(concluido)


def main():
    while True:
        exibir_menu()

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            adicionar_tarefa()
        elif opcao == "2":
            exibir_tarefas()
        elif opcao == "3":
           remover_tarefas()
        elif opcao == "4":
           remover_tarefas_concluidas()

        elif opcao == "5":
           concluir_tarefa()
        elif opcao == "6":
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
